```python
import ollama
from datasets import load_dataset
import json


from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

from nltk.translate.meteor_score import meteor_score



try:
    dataset = load_dataset("wmt14", "fr-en", split="test[:100]")
except Exception as e:
    print(f"Failed to load dataset: {e}")
    print("Please ensure you have an active internet connection and the 'datasets' library is installed correctly.")
    exit()


# --- 2. Initialize Model and Scorers ---

MODEL_NAME = 'gemma3:1b'
print(f"Using Ollama model: {MODEL_NAME}")
print("-" * 20)

# Initialize smoothing function for BLEU
smoothie = SmoothingFunction().method1

# Lists to store all references and generated candidates for final scoring
all_references = []
all_candidates = []
meteor_scores = []
bleu_scores_per_sentence = []

# List to store detailed results for each sample
detailed_results = []


# --- 3. Iterate Through Dataset and Evaluate ---

print(f"Evaluating on {len(dataset)} samples from the WMT14 fr-en dataset...")

for item in dataset:
    # The source language (German) text will be our prompt
    prompt = item['translation']['fr']

    # The target language (English) text is our reference
    reference_text = item['translation']['en']

    # Tokenize the reference and candidate


    structured_prompt = f"""Translate the following French text to English. Your response should contain only the translated text.

    ### Example
    French: Reprise de la session
    English: Resumption of the session

    ### Task
    French: {prompt}
    English:"""

    OLLAMA_OPTIONS = {
    "temperature": 0.2,
        }

    try:
        # Generate a translation using the Ollama model
        response = ollama.generate(
            model=MODEL_NAME,
            prompt=structured_prompt,
            stream=False,
            options=OLLAMA_OPTIONS
        )

        # Extract the generated text
        candidate_text = response['response'].strip()

        # Append to our lists for overall scoring
        all_references.append([reference_text]) # sacrebleu expects a list of lists of references
        all_candidates.append(candidate_text)

        # Calculate NLTK BLEU for this specific item
        reference_tokenized_single = [nltk.word_tokenize(reference_text)]
        candidate_tokenized_single = nltk.word_tokenize(candidate_text)
        current_bleu = sentence_bleu(reference_tokenized_single, candidate_tokenized_single, smoothing_function=smoothie)
        bleu_scores_per_sentence.append(current_bleu)

        # Calculate NLTK METEOR for this specific item
        reference_tokenized_meteor = nltk.word_tokenize(reference_text)
        current_meteor = meteor_score([reference_tokenized_meteor], candidate_tokenized_single)
        meteor_scores.append(current_meteor)

        print(f"Prompt (fr): {prompt}")
        print(f"Reference (en): {reference_text}")
        print(f"Candidate (en): {candidate_text}")
        print(f"BLEU (NLTK) for this sentence: {current_bleu:.2f}")
        print(f"METEOR for this sentence: {current_meteor:.2f}")
        print("-" * 10)

        # Store detailed results for this sample
        detailed_results.append({
            "prompt": prompt,
            "reference": reference_text,
            "candidate": candidate_text,
            "bleu_nltk_sentence": current_bleu,
            "meteor_sentence": current_meteor
        })

    except Exception as e:
        print(f"Error processing item: {e}")
        continue


# --- 4. Calculate and Display Final Scores ---

if not all_candidates:
    print("No candidates were generated. Exiting.")
    exit()

# Calculate corpus-level BLEU score
# sacrebleu's corpus_bleu is designed for this and is the standard way to report BLEU.

avg_bleu_nltk = sum(bleu_scores_per_sentence) / len(bleu_scores_per_sentence) if bleu_scores_per_sentence else 0

# Calculate average METEOR scores
avg_meteor = sum(meteor_scores) / len(meteor_scores) if meteor_scores else 0


print("\n--- Overall Evaluation Results ---")
print(f"Total samples evaluated: {len(all_candidates)}")
print(f"Average NLTK BLEU Score: {avg_bleu_nltk * 100:.2f}")
print(f"Average METEOR Score: {avg_meteor * 100:.2f}")
print("---------------------------------")

# --- 5. Save Results to JSON File ---
results_to_save = {
    "overall_results": {
        "total_samples_evaluated": len(all_candidates),
        "average_nltk_bleu_score": avg_bleu_nltk * 100,
        "average_meteor_score": avg_meteor * 100
    },
    "detailed_sample_results": detailed_results
}

# Clean up MODEL_NAME to be a valid filename
filename = f"{MODEL_NAME.replace(':', '_').replace('/', '_')}_evaluation_results.json"

try:
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(results_to_save, f, ensure_ascii=False, indent=4)
    print(f"Evaluation results saved to {filename}")
except Exception as e:
    print(f"Failed to save results to JSON file: {e}")
```
